# Introduction

A fifteen-year-old child in Gaza has been the victim of five major military operations in their lifetime. They have seen their friends and family killed, their homes destroyed, and their schools bombed. They have lived through a blockade that has lasted for over a decade, leaving little hope for the future. These wars, fought between Israel and a Hamas government elected before they were born, has continued to make world headlines, but often leaving their experiences unseen and unheard. This phenomenon, what Gene Demby calls the denial of the "right to a story" ([Demby, 2023](https://www.npr.org/transcripts/1197954284)) has garnered increased scrutiny over the past years, particularly as existence as a non-Israeli in the West Bank and Gaza has become deadlier. In May, the United Nations warned that 2023 would be the deadliest year for Palestinians in recent memory ([United Nations, 2023](https://press.un.org/en/2023/sc15293.doc.htm)). Even earlier, in February, Al Jazeera reported that “With 62 Palestinians, including 13 children, killed this year so far, and hundreds of others injured, the first two months of 2023 have been the deadliest compared with the same period since 2000” ([Tahhan, 2023](https://www.aljazeera.com/news/2023/2/23/catastrophic-palestinians-recount-deadly-israeli-raid-on-nablu)). This coming on the heels of 2022, which marked the single deadliest year for Palestinians since the Second Intifada in 2005 ([UNSC, 2023](https://www.un.org/unispal/document/with-2022-deadliest-year-in-israel-palestine-conflict-reversing-violent-trends-must-be-international-priority-middle-east-coordinator-tells-security-council-press-release-sc-15179/)). 

Unsurprisingly, the occupation of Palestine has become increasingly unpopular in global opinion. A 2013 BBC poll ranked Israel as the fourth most disliked nation on the planet ([BBC, 2013](https://www.bbc.com/news/world-europe-22624104)), with many citing illegal settlements ([Beauchamp, 2013](https://www.vox.com/2018/11/20/18080086/israel-palestine-global-opinion)) on the West Bank and the ongoing Gazan blockade which continues to deny basic necessities to its inhabitants. There remains a gap, however, in popular opinion and most western reporting on the region. The purpose of this large-data analysis is to quantitatively assess the erasure of Palestinians in mainstream western reporting of the conflict.  

Precipitated by global reaction to the violent and tragic Hamas attack on October 7, 2023, killing an estimated 1,200 Israeli civilians and soldiers ([Frayer, 2023](https://www.npr.org/2023/11/11/1212458974/israel-revises-death-toll-hamas-attacks-oct-7)), this report aims to compare reporting on the 2014 Gaza war to that of the ongoing conflict to identify shifts in how western media companies tell the stories of Palestinians on the ground. Two trends seem likely: 1) that reporting in 2023 has shifted focus to the repercussions of such wars as punishment inflicted upon civilians to mirror public sentiment, or 2) that in reaction to the unprecedented brutal nature of the Hamas attack compounded by western powers’ redoubled military support for Israel, reporting will continue to largely ignore the Palestinian experience.  

**Hypothesis<sub>0</sub>**: Global sympathy towards Palestinian liberation does not significantly influence or correlate with heightened media coverage of Palestinian perspectives during the 2023 Gaza war compared to that of 2014.  

**Hypothesis<sub>A</sub>**: Increasing global sympathy towards Palestinian liberation will positively correlate with heightened media coverage of Palestinian perspectives during the 2023 Gaza war compared to that of 2014.

While both hypotheses have their merits, recency bias favors the null hypothesis (a major factor in deciding to conduct this experiment). So too do instances of respected media outlets unquestionably relaying misinformation and disinformation at the start of the 2023 war. For instance, both CNN and BBC aired a now debunked claim that 40 babies had been beheaded in Kibbutz Kfar Aza ([Swann, 2023](https://www.politifact.com/article/2023/nov/21/israel-hamas-war-what-we-know-about-beheaded-babie/)). A claim that has since been retracted by most major outlets, with CNN’s Sara Sidner remarking "Yesterday the Israeli Prime Minister's office said that it had confirmed Hamas beheaded babies and children while we were live on the air. The Israeli government now says today it CANNOT confirm babies were beheaded. I needed to be more careful with my words and I am sorry" ([Lohani, 2023](https://www.tbsnews.net/features/panorama/beheaded-babies-and-other-lessons-controlling-israel-palestine-narrative-720270)). While many are frustrated at the recent level of reporting, it is important to find the truth in the data and contextualize this with the same type of reporting in previous crises. It is only then that we will be able to fasten hope—or fear—of the Palestinian’s right to their story.

Using Sherri's code as a starting point, I supplemented it with Jeremy DiBattista's code to create a sentiment analysis function that would return a dataframe with the sentiment score for each article ([DiBattista, 2021](https://towardsdatascience.com/the-best-python-sentiment-analysis-package-1-huge-common-mistake-d6da9ad6cdeb)). Another telling report from the NLTK analysis is by use of the `concordance` and `common contexts` functions. While `concordance` returns the sentences in which the word appears, `common contexts` returns the words that appear in the same context as the word in question. For instance, in the 2014 dataset, the word "Palestinian" appears in the context of "Palestinian militants", "Palestinian civilians", "Palestinian territory", and "Palestinian Authority". In the 2023 dataset, strikingly, the word "Palestinian" does not appear in the context of "Palestinian militants", suggesting that perhaps clearer lines have been drawn in reporting between the Palestinian people and Hamas. However, the word "Palestinian" does appear in the context of "Palestinian civilians", "Palestinian Authority", and "Palestinian state". Another striking difference between the two datasets is the inclusion of "Palestinian terrorism" in the 2014 with no such instances in the 2023 sample, further highlighting a rhetorical separation between Hamas and the Palestinian people writ whole.

In the case of the `common contexts` outputs, both the 2014 and 2023 datasets show heavy use of the terms "militants" in instances where both the words "Palestinians" and "Hamas" appear together. What is surprising is the lack of common context between the words "Israeli", "settler", "terror", "terrorism", and "attack" in 2014. This changes in 2023, where some tenuous linkages are identified between "Israeli", "settler", and "terror". One can assume that as the Israeli settler program in the West Bank ramps up, so too has the media coverage surrounding it.

Top 30 most common words from both datasets. The wordcloud here is particularly helpful in visualizing how few times "Palestinians" are referenced in the corpus.

```python
#Taken from in-class tutorial
text_lower_2014 = [t.lower() for t in nltk_2014 if t.isalnum()]
text_stops_2014 = [t for t in text_lower_2014 if t not in stops]

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
    
text_clean_2014 = [WordNetLemmatizer().lemmatize(t, get_wordnet_pos(t)) for t in text_stops_2014]
```

```python
dist_2014 = FreqDist(text_clean_2014)
dist_2014.plot(30, cumulative=False)
dist_2014.most_common(30)
```

```python
!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = ' '.join(text_clean_2014)

wordcloud = WordCloud().generate(text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### TF-IDF and its associated heatmaps

Perhaps the most elucidating method used in this analysis is via TF-IDF. The TF-IDF formula identifying the importance of a word in the text highlights how little Palestinians are mentioned in the corpus. In the heatmaps associated with both datasets, I've chosen to highlight the words "Palestinians" and "Civilians" (indicated by a red dot), illustrating the scant number of times they appear in the corpus. Of all the tools used in this analysis, TF-IDF does well illustrate that our important key words are not just few in frequency, but also in importance.

```python
!pip install sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

pd.options.display.max_rows = 1000
from pathlib import Path  
import glob
```

## Conclusion

### Failure to reject the null hypothesis

As illustrated by these results, the place held by Palestinians in mainstream western media are strikingly similar from 2014 to 2023. The frequency and importance of the Palestinian people remain almost identical with the only real observable difference being that 2014 rhetoric had a higher propensity to conflate “Palestinians” with “Militants” and “Terrorists”. Although the originally planned sentiment analysis fell short, the tools utilized here provides a compelling argument that even if sentiment is generally positive, the lack of representation could be an even bigger hurdle in the struggle for a “right to a story”.  

Some considerable drawbacks came with choosing such a large dataset, including a diminished applicability of sentiment analysis models. While the large _n_ experiment lent itself well to aggregation, TF-IDF, and NER, future sentiment analysis should be considered with working with a far more curated set of articles. Another possibility for future study would be the use of time-series analysis to track how the Palestinian story evolves as the war drags on. Intuition indicates that in the 2023 case, media reports would be far less focused on Palestinian civilians in the immediate aftermath of October 7, with a gradual shift occurring as Palestinian civilian casualties skyrockets paired with yet another humanitarian crisis in Gaza. Finally, there remains unexplored a comparison between western legacy media and reporting from the global south. Especially given that at least 64 journalists have been killed in Gaza since October 7 ([Martin, 2023](https://www.npr.org/2023/12/19/1220210900/number-of-journalists-killed-in-gaza-since-oct-7-attacks-called-unprecedented-lo#:~:text=Transcript-,At%20least%2064%20journalists%20have%20been%20killed%20in%20Gaza%20since,Journalists%2C%20about%20the%20unprecedented%20loss.)), the vast majority of which  having been of Middle Eastern descent. The obvious challenge here is conducting quantitative text-analysis with non-English sources.
